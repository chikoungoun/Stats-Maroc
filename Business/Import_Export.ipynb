## import - Export

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Evolution of days needed for import operations (from 2005 to 2017)**

```python
imp_time = pd.read_csv('time_import_days.csv',index_col=0)
imp_time.sort_index(ascending=True,inplace=True)
imp_time.head()
```

```python
imp_time.plot.line(figsize=(13,6),color='red',fontsize=14)
```

**Evolution of the Number of documents needed to import goods (from 2005 to 2014)**

```python
imp_docs = pd.read_csv('number_docs_import.csv')
imp_docs.sort_index(ascending=False,inplace=True)
imp_docs.head()
```

```python
imp_docs.plot.line(figsize=(13,6),color='purple',fontsize=14)
```

**Evolution of days needed for export operations (from 2005 to 2017)**

```python
exp = pd.read_csv('time_export_days.csv',index_col=0)
exp.sort_index(ascending=True,inplace=True)
exp.head()
```

```python
exp.plot.line(figsize=(13,6),color='green',fontsize=14)
```
